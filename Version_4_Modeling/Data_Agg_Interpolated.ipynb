```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
sepsis = pd.read_csv('Sepsis_Compiled_Full.csv',index_col=0)
```

```python
sepsis['HospAdmTime'] = abs(sepsis['HospAdmTime'])
sepsis['Age'] = [int(foo) for foo in sepsis['Age']]
```

```python
patient_frames = []
i = 0
for a,b in sepsis.groupby('PatientID'):
    if i % 700 == 0:
        print(i, ' th pass')
    b = b.interpolate(method='linear').fillna(0)
    assert np.sum(pd.isnull(b.values)) == 0, 'There are still NaNs!'
    patient_frames.append(b)
    i += 1
    
    
```

```python
sepsis = pd.concat(patient_frames)
```
